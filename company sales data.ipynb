```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = pd.read_csv('company_sales_data.csv') 
```

```python
plt.xlabel('Month number')
plt.ylabel('Total profit')

x = data['month_number'].to_numpy()
y = data['total_profit'].to_numpy()

plt.plot(x, y)
plt.xticks(x)
plt.title('Company profit per month')
```

```python
plt.xlabel('Month number')
plt.ylabel('Total profit')
plt.title('Company sale data of last year')
plt.plot(x, y, ls='dotted', c='r', marker='o', mfc='black', lw=3, label='Profit data of last year')
plt.xticks(x)
plt.legend(loc='lower right',)
```

```python
plt.xlabel('Month number')
plt.ylabel('Sales unit in number')
z1 = data['facecream'].to_numpy()
z2 = data['facewash'].to_numpy()
z3 = data['toothpaste'].to_numpy()
z4 = data['bathingsoap'].to_numpy()
z5 = data['shampoo'].to_numpy()
z6 = data['moisturizer'].to_numpy()

plt.plot(x, z1, marker='o', lw=3, label='face cream sales data')
plt.plot(x, z2, marker='o', lw=3, label='face wash sales data')
plt.plot(x, z3, marker='o', lw=3, label='Tooth paste sales data')
plt.plot(x, z4, marker='o', lw=3, label='Bathing soap sales data')
plt.plot(x, z5, marker='o', lw=3, label='Shampoo sales data')
plt.plot(x, z6, marker='o', lw=3, label='Moisturizer sales data')
plt.xticks(x)
plt.title('Sales data')
plt.legend()
```

```python
plt.xlabel('Month number')
plt.ylabel('Number of units sold')
plt.scatter(x,z3,label='Tooth paste Sales data')
plt.title('Tooth paste Sales data')
plt.xticks(x)
plt.grid(ls='--')
plt.legend()
```

```python
import numpy as np

months = data['month_number'].to_numpy()

plt.bar(months - 0.2, z1, width=0.4, label='Face Cream sales data')

plt.bar(months + 0.2, z2, width=0.4, label='Face Wash sales data')

plt.xlabel('Month Number')
plt.ylabel('Sales units in number')
plt.title('FaceWash and FaceCream sales data')
plt.xticks(x)
plt.legend(loc='upper left')
plt.grid(ls='--')
```

```python
z = ['FaceCream', 'FaseWash', 'ToothPaste', 'Bathing soap', 'Shampoo', 'Moisturizer']
SaleData   = [data['facecream'].sum(), data['facewash'].sum(), data['toothpaste'].sum(), 
         data['bathingsoap'].sum(), data['shampoo'].sum(), data['moisturizer'].sum()]
plt.axis("equal")
plt.pie(SaleData, labels=z, autopct='%1.1f%%')
plt.legend( loc ='lower right')
plt.title('Sales data')
```
